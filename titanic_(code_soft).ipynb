```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (assuming you've already done the preprocessing steps)
data = pd.read_csv('titan.csv')

# Count the number of survivors and non-survivors
survived_counts = data['Survived'].value_counts()

# Create a pie chart
labels = ['Did Not Survive', 'Survived']
plt.pie(survived_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Survival Distribution on the Titanic')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (assuming you've already done the preprocessing steps)
data = pd.read_csv('titan.csv')

# Group the data by gender and count survivors
survived_by_gender = data.groupby('Sex')['Survived'].value_counts()

# Create a bar chart
labels = ['Female - Did Not Survive', 'Female - Survived', 'Male - Did Not Survive', 'Male - Survived']
plt.bar(labels, survived_by_gender.values)
plt.xlabel('Gender and Survival')
plt.ylabel('Number of Passengers')
plt.title('Survival Distribution on the Titanic by Gender')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (assuming you've already done the preprocessing steps)
data = pd.read_csv('titan.csv')

# Filter data for male passengers, using the numeric value for male (0)
male_data = data[data['Sex'] == 0]

# Count survival for males
male_survived_counts = male_data['Survived'].value_counts()

# Create a pie chart for male survival
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.subplot(1, 2, 1)  # Subplot for male survival
plt.pie(male_survived_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%', startangle=90)
plt.title('Male Survival Distribution')
plt.axis('equal')

# Filter data for female passengers, using the numeric value for female (1)
female_data = data[data['Sex'] == 1]

# Count survival for females
female_survived_counts = female_data['Survived'].value_counts()

# Create a pie chart for female survival
plt.subplot(1, 2, 2)  # Subplot for female survival
plt.pie(female_survived_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%', startangle=90)
plt.title('Female Survival Distribution')
plt.axis('equal')

plt.tight_layout()  # Ensures labels do not overlap
plt.show()
```
