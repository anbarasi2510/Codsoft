```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Sample data
data = pd.DataFrame({
    'Name': ['Movie1', 'Movie2', 'Movie3', 'Movie4', 'Movie5'],
    'Year': [2019, 2021, 2019, 2010, 2021],
    'Duration': [109, 90, 110, 105, 90],
    'Genre': ['Drama', 'Drama, Musical', 'Comedy, Romance', 'Drama', 'Drama'],
    'Rating': [7.0, None, 4.4, None, 5.0],
    'Votes': [8, None, 35, None, 50],
    'Director': ['Gaurav Bakshi', 'Soumyajit Majumdar', 'Ovais Khan', 'Amol Palekar', 'Unknown'],
    'Actor 1': ['Rasika Dugal', 'Sayani Gupta', 'Prateik', 'Rajat Kapoor', 'Unknown'],
    'Actor 2': ['Vivek Ghamande', 'Plabita Borthakur', 'Ishita Raj', 'Rituparna Sengupta', 'Unknown'],
    'Actor 3': ['Arvind Jangid', 'Roy Angana', 'Siddhant Kapoor', 'Antara Mali', 'Unknown']
})

# Handle missing values
data = data.dropna(subset=['Rating'])
data = data.fillna('Unknown')

# Encode categorical features
encoder = OneHotEncoder(handle_unknown='ignore')
encoded_features = encoder.fit_transform(data[['Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3']]).toarray()
encoded_feature_names = encoder.get_feature_names_out(['Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3'])

# Combine encoded features with numerical features
features = np.concatenate([encoded_features, data[['Year', 'Duration', 'Votes']].values], axis=1)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(features, data['Rating'].values, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train a regression model (e.g., Random Forest)
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print('MAE:', mean_absolute_error(y_test, y_pred))
print('MSE:', mean_squared_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))

# Plot feature importances
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 8))
plt.title("Feature Importances")
plt.bar(range(X_train.shape[1]), importances[indices], color="r", align="center")
plt.xticks(range(X_train.shape[1]), np.array(encoded_feature_names.tolist() + ['Year', 'Duration', 'Votes'])[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()
```

```python
# Plot feature importances as a pie chart
importances = model.feature_importances_
feature_names = np.array(encoded_feature_names.tolist() + ['Year', 'Duration', 'Votes'])

# Create a DataFrame for better handling of feature importances
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})

# Sort the DataFrame by importance
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Plot pie chart
plt.figure(figsize=(10, 10))
plt.pie(importance_df['Importance'], labels=importance_df['Feature'], autopct='%1.1f%%', startangle=140)
plt.title("Feature Importances (Pie Chart)")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('IMDb Movies India.csv')

# Handle missing values in the 'Rating' column
data = data.dropna(subset=['Rating'])

# Convert the Rating column to numeric
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')

# Function to plot top 10 bar diagram for a given feature
def plot_top_10_by_feature(data, feature, title):
    top_10 = data.groupby(feature)['Rating'].mean().sort_values(ascending=False).head(10)
    plt.figure(figsize=(12, 8))
    plt.barh(top_10.index, top_10.values, color='skyblue')
    plt.xlabel('Average Rating')
    plt.ylabel(feature)
    plt.title(title)
    plt.gca().invert_yaxis()  # To display the highest rated at the top
    plt.show()

# Plot top 10 highest-rated movies
top_10_movies = data.sort_values(by='Rating', ascending=False).head(10)
plt.figure(figsize=(12, 8))
plt.barh(top_10_movies['Name'], top_10_movies['Rating'], color='skyblue')
plt.xlabel('Rating')
plt.ylabel('Movie Name')
plt.title('Top 10 Highest-Rated Movies')
plt.gca().invert_yaxis()
plt.show()

# Plot top 10 actors by average rating
plot_top_10_by_feature(data, 'Actor 1', 'Top 10 Actors by Average Rating (Actor 1)')
plot_top_10_by_feature(data, 'Actor 2', 'Top 10 Actors by Average Rating (Actor 2)')
plot_top_10_by_feature(data, 'Actor 3', 'Top 10 Actors by Average Rating (Actor 3)')

# Plot top 10 directors by average rating
plot_top_10_by_feature(data, 'Director', 'Top 10 Directors by Average Rating')

# Plot top 10 genres by average rating
plot_top_10_by_feature(data, 'Genre', 'Top 10 Genres by Average Rating')
```
